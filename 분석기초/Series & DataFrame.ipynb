#### Series: 1차원 구조

```python
index = ['apple','banana','orange','kiwi']
data = [10, 9, 5, 3]

#series = pd.Series(index=index, data)  # Error
series = pd.Series(data, index=index)  # OK
print(series)
print()

## Filtering :
conditions = [True, False, False, True]
print(series[conditions])
print()

## Sorting :
print(series.sort_index())
print()
print(series.sort_values())
```

```python
## 파일 읽어오기 :
data = pd.read_csv("data/weight_log.csv", index_col = "회차")  #추가 주요옵션: encoding=utf-8/cp949, header=[0,1]
data
```

```python
print(data.columns)
print(data.index)
data.head()
data.tail()
```

```python
print(data.describe())
print(data["몸무게"].mean())
print(data["지점"].unique())
print(data["지점"].value_counts())                 # Unique 갯수 표시
print(data["지점"].value_counts(normalize = True)) # Unique 비율표시
```

```python
data["지점"] = data["지점"].replace("강남구", "강남").replace('서초구','서초')
data['몸무게(fill)'] = data['몸무게'].fillna(data["몸무게"].mean())
data
```

```python
print(data['측정일'])
data['측정일'] = pd.to_datetime(data['측정일'])
print(data['측정일'])
data['측정일'].dt.month
```

```python
## 여러개 컬럼 기준으로 값 읽어보기 :
data[["이름", "측정일", "몸무게"]]
```

```python
data.loc[7, ["이름", "측정일", "몸무게"]]
```

```python
## 컬럼값 조건에 따른 값 :

data[data["담당"] == "박길동"]
data[data["담당"].isin(["박길동", "김길동"])]
data[data["담당"].str.contains("김")]
data[data["몸무게"] > 75]
data[(data["몸무게"] > 75.5) & (data["담당"] == "김길동")]
```

```python
data[~data["몸무게"].isnull()]
```

```python
# 또는 
data[data["몸무게"].notnull()]
```

```python
data.loc[data["담당"] == "김길동", "몸무게"]
data.loc[data["담당"] == "김길동", ["이름", "측정일", "몸무게"]]
```

```python
## pivot_table :
pd.pivot_table(data, index=["지점", "담당"], values="몸무게", aggfunc='sum') #aggfunc 기본값: mean평균
```

```python
## Column 추가 예시 :
data["몸무게(gram)"] = data["몸무게"] * 100
data[["몸무게", "몸무게(gram)"]]

data.loc[data["몸무게"] > 75.6, "상태"] = "비만"
data.loc[data["몸무게"] <= 75.6, "상태"] = "정상"
#data[["몸무게", "비만", "상태"]]
data
```
