```python
### 대체 (replace)

# '강남구'라고 적힌 값을 '강남'으로 통일해주기 위해 replace를 사용
# 이를 '지점' 컬럼에 다시 대입

data["지점"] = data["지점"].replace("강남구", "강남")

# 서초도 마찬가지
data["지점"] = data["지점"].replace('서초구','서초')

data
```

```python
### 결측치 (fillna)

# 몸무게가 비어있는(NaN) 데이터에 한해서 75.8 kg로 채워준다

# 이를 '몸무게(fill)'이라는 새로운 컬럼에 추가
data['몸무게(fill)'] = data['몸무게'].fillna(75.8)

data
```

```python
### shift

# .shift(+1)으로 이전 날짜의 몸무게를 가져옴
# 이전 기록이 없다면(=첫 번째 데이터) NaN을 가져옴
data['몸무게(fill)(prev)'] = data['몸무게(fill)'].shift(+1)

# .shift(-1)으로 다음 날짜의 몸무게를 가져옴
# 다음 기록이 없다면(=마지막 데이터) NaN을 가져옴
data['몸무게(fill)(next)'] = data['몸무게(fill)'].shift(-1)

data
```

```python
data['몸무게(fill)']-data['몸무게(fill)(prev)']
```

하지만 현실적으로, 판다스에서 컬럼을 추가할 때 가장 많이 하는 것은 조건마다 컬럼을 추가하는 것입니다. 가령 "A라는 조건에는 1이라는 값을 넣어주고, B라는 조건에는 2라는 값을 넣어준다."라는 방식을 사용하는 경우가 많습니다.

이렇게 조건별로 컬럼을 추가해주는 방식은 크게 두 가지로 나뉠 수 있는데,
 1. 첫 번째는 각 조건마다 컬럼을 하나씩 추가하는 방법
 2. 두 번째는 컬럼은 하나인데 그 안에 조건마다 값이 다르게 들어가는 방법입니다.
 - 먼저 첫 번째 방식으로 컬럼을 추가해보겠습니다.

```python
# 몸무게가 75.6 kg보다 크면 True이고, 그렇지 않으면 False
data["비만"] = data["몸무게"] > 75.6

data[["몸무게", "비만"]]
```

```python
data.loc[data["몸무게"] > 75.6, "상태"] = "비만"
data.loc[data["몸무게"] <= 75.6, "상태"] = "정상"

data[["몸무게", "비만", "상태"]]
```

```python
data
```

```python
# 몸무게가 75.5 kg 미만이면 True, 아니면 False
data["목표초과"] = data["몸무게"] < 75.5

# 몸무게가 75.5 kg 이상이고 76.0 kg 미만이면 True, 아니면 False
data["목표달성"] = (data["몸무게"] >= 75.5) & (data["몸무게"] <= 76.0)

# 몸무게가 76.0 kg 초과면 True, 아니면 False
data["목표실패"] = data["몸무게"] > 76.0

data[["몸무게", "목표초과", "목표달성", "목표실패"]]
```

### apply 함수

```python
def band_weight(weight):
    # weight 값이 NaN이면 NaN 그대로
    if pd.isnull(weight):
        return np.nan

    # weight 값이 75.5 kg 미만이면 '초과'
    if weight < 75.5:
        return '초과'
    
    # weight 값이 75.5 kg 이상이고 76.0 kg 이하면 '달성'
    elif weight >= 75.5 and weight < 76.0:
        return '달성'
    
    # 모든 조건에 해당하지 않으면 '미달'
    else:
        return '미달'
```

```python
# '몸무게' 컬럼에 apply(band_weight) 실행
data["목표"] = data["몸무게"].apply(band_weight)
```

```python
data1 = data.copy()
data1["목표"] = data1["몸무게"].apply(band_weight)
data1
```
