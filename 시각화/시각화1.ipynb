## BarPlot

```python
plt.figure(figsize=(18, 3))
sns.barplot(data=data, x='연령대코드(5세단위)', y='수축기혈압') # 수축기혈압의 평균을 표시
```

```python
import numpy as np

plt.figure(figsize=(18, 3))
sns.barplot(data=data, x='연령대코드(5세단위)', y='수축기혈압', estimator=np.sum) # 총합
```

```python
plt.figure(figsize=(18, 3))
sns.barplot(data=data, y='연령대코드(5세단위)', x='수축기혈압') # x, y 를 바꿔서 그림
```

```python
plt.figure(figsize=(3, 9))
sns.barplot(data=data, y='연령대코드(5세단위)', x='수축기혈압', orient='h') # x, y 를 바꿔서 그려볼 경우, 세로로 표현
```

```python
plt.figure(figsize=(16, 4))
sns.barplot(data=data, x='연령대코드(5세단위)', y='수축기혈압', estimator=np.mean, ci=False) # ci = False : 신뢰구간 표시안함
```

```python
plt.figure(figsize=(16, 4))
sns.barplot(data=data, x='연령대코드(5세단위)', y='수축기혈압', estimator=np.sum, ci='sd') # ci = sd : 표준편차
```

## BoxPlot

- 최소, 중앙(median), 최대, 1, 3 사분위 수 등을 표현함
- 점(diamond형) 은 이상치 outlier 의미

1) 최소값
2) 제 1사분위수
3) 제 2사분위수, 즉 중앙값
4) 제 3사분위수
5) 최댓값

### 분위수 (Quantile)

- 중앙값이 데이터를 2개의 동일한 사이즈로 나누었을 때, 그 경계값을 의미한다면, 4분위수는 데이터를 4개의 동일한 사이즈로 나누었을 때 그 경계값들을 의미함

- 자료의 1/4 (25%) 지점 : 1분위수
- 자료의 2/4 (50%) 지점 : 2분위수 (=중앙값)
- 자료의 3/4 (75%) 지점 : 3분위수
